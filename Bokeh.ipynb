```python
from bokeh.layouts import widgetbox, column
from bokeh.models import Slider, ColumnDataSource, Select
from bokeh.plotting import figure
from numpy.random import random
def bkapp(doc):
    year = 1970
    data_a = data[data["Year"] == year]
    
    #source = ColumnDataSource(data_FLC)
    data_FLC = data_a
    source = ColumnDataSource(data_FLC)
    unique_region = list(data["region"].unique())
    mapper = CategoricalColorMapper(
            factors = unique_region,
            palette =['red', 'green', 'blue','yellow','grey','black'])
    hover = HoverTool(#tooltips = none,
                  tooltips = {'fertility':'@fertility', 'life':'@life', 'country':'@Country','child_mortality':'@child_mortality','gdp':'@gdp'}, 
                  #mode = 'hline'
                 )
    # create plot
    plot = figure(x_axis_label = 'fertility', y_axis_label = 'life',tools = [hover, 'crosshair'],title = 'Plot for the year: 1970' )
    plot.circle('fertility', 'life', 
         size = 10, 
         source= source,
         color = {'field':'region', 'transform': mapper},
         hover_color = 'gold',
        )
    

    # add slider
    slider = Slider(start = 1970, end = 2010, value =year, step =1, title = 'Year')
    menu1 = Select(options = ['fertility', 'life', 'child_mortality','gdp' ],
                  value = 'fertility', title = 'Value_1')
    menu2 = Select(options = ['fertility', 'life', 'child_mortality','gdp'],
                  value = 'life', title = 'Value_2')
    # add callback to widget
    def update_year(attr, old, new):
        year = slider.value 
        value1 = menu1.value 
        value2 = menu2.value
        data_a = data[data["Year"] == year]
        a = data_a.copy()
        a['fertility'] = data_a[value1]
        a['life'] = data_a[value2]
        data_FLC = a
        source.data = data_FLC
        new_title = 'Plot for the year: ' + str(year)
        plot.title.text = new_title
        plot.xaxis.axis_label = str(value1)
        plot.yaxis.axis_label = str(value2)
        
    def callback1(attr, old, new):
        value1 = menu1.value 
        value2 = menu2.value
        year = slider.value
        data_a = data[data["Year"] == year]
        a = data_a.copy()
        a['fertility'] = data_a[value1]
        a['life'] = data_a[value2]
        data_FLC = a
        source.data = data_FLC
        new_title = 'Plot for the year: ' + str(year)
        plot.title.text = new_title
        plot.xaxis.axis_label = str(value1)
        plot.yaxis.axis_label = str(value2)
        
    def callback2(attr, old, new):
        value1 = menu1.value
        value2 = menu2.value
        year = slider.value
        data_a = data[data["Year"] == year]
        a = data_a.copy()
        a['fertility'] = data_a[value1]
        a['life'] = data_a[value2]
        data_FLC = a
        source.data = data_FLC
        new_title = 'Plot for the year: ' + str(year)
        plot.title.text = new_title
        plot.xaxis.axis_label = str(value1)
        plot.yaxis.axis_label = str(value2)
        
       
        
    slider.on_change('value', update_year)   # we want the callback to be executed when value of slider changes
    menu1.on_change('value', callback1)
    menu2.on_change('value', callback2)
    # arrange plots and widgets
    layout = column(slider,menu1,menu2, plot)
    
    # add to document
    doc.add_root(layout)

show(bkapp)
```
